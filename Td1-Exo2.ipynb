```python
# Which person had an average velocity of 0 during their life ? e.g. died where they where born .
# death[(death.lieunaiss==death.lieudeces)]
# How many people died in the same place they were born by communes ?
death[(death.lieunaiss==death.lieudeces)].groupby("commnaiss").count().sort_values(by=['nomprenom'],ascending=False)[:20]
```

```python
# From where were most of the people dying in 2018 ?
death.groupby("paysnaiss").count().sort_values(by=['nomprenom'],ascending=False)[:10]
```

```python
# What is the top 10 of the most dangerous places ?
death.groupby("lieudeces").count().sort_values(by=['nomprenom'],ascending=False)[:10]
```

```python
# How many death each day during the month of January by country?
plt.rcParams['figure.figsize'] = [20, 20] # Taille du graphique en pouces
plt.rcParams['figure.dpi'] = 200 # résolution en points par pouce dot per inches

total_death=death[(death.datedeces<20180200)&(death.datedeces>20180100)].pivot_table('nomprenom',index='datedeces',columns='paysnaiss',aggfunc=np.count_nonzero)
total_death.plot()

# death[(death.datedeces>=20180000)].groupby("datedeces").nomprenom.count().plot()
plt.grid()
```

```python
# What is the average age at death by sexe ?
death['age']=round(death.datedeces/10000-death.datenaiss/10000)
total_death=death[(death.age>0)&(death.age<150)].pivot_table('nomprenom',index='age',columns='sexe',aggfunc=np.count_nonzero)
total_death.plot()
plt.grid()
```

```python
# What is the average age at death in the 4 country of Algeria , Moroco , Italia and Espagna?
death['age']=round(death.datedeces/10000-death.datenaiss/10000)
total_death=death[(death.age>0)&(death.age<150)&((death.paysnaiss=="ALGERIE")|(death.paysnaiss=="MAROC")|(death.paysnaiss=="ITALIE")|(death.paysnaiss=="ESPAGNE"))].pivot_table('nomprenom',index='age',columns='paysnaiss',aggfunc=np.count_nonzero)
total_death.plot()
plt.grid()
```
